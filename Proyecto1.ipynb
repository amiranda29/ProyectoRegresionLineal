2. Crear la clase que hara la regresion lineal

```python
class RegresionLineal():
    #Inicializar parametros
    def __init__(self,x,y,epoch,impr_error,learning_rate):
        self.x=x
        self.y=y
        self.epoch=epoch
        self.impr_error=impr_error
        self.learning_rate=learning_rate
        self.n, self.m= x.shape
        self.w=np.zeros((self.n,1))
        self.b=0
        #Actualizar los pesos del modelo
        for i in range(self.epoch):
            self.actualizarPesos()
    
        #Funcion para actualizar pesos
    def actualizarPesos(self):
        predecirY = self.predecir(self.x)

        #Calculo de gradientes
        dw = -(self.x.T).dot(self.y - predecirY)/self.m
        db = -np.sum(self.y - predecirY)/self.m

        #Actualizar los pesos
        self.w = self.w - self.learning_rate*dw
        self.b = self.b - self.learning_rate*db
    
        #Funcion para predecir precios de venta 
    def predecir(self,x):
        return x.dot(self.w) + self.b
    
    def imprimirPesos(self):
        print('Los pesos son: ')
        print(self.w)
        print()
        print('Error en el modelo: ', self.b)
```
